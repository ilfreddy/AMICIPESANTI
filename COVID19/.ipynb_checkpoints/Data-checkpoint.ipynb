## Open a CSV file using Pandas

```python
import pandas
df = pandas.read_csv('data.csv')
df.head(100)
```

## Read a GeoJSON file

```python
x=df.deceduti
y=df.Incremento
```

```python
import numpy as np
Polynomial = np.polynomial.Polynomial
```

```python
pfit, stats = Polynomial.fit(x,y, 1, full=True)
```

```python
print('Raw fit results:', pfit, stats, sep='\n')
```

```python
import pylab
x0, m = pfit
resid, rank, sing_val, rcond = stats
rms = np.sqrt(resid[0]/len(x))

print('Fit: Y = {:.3f}[X] + {:.3f}'.format(m, x0),
      '(rms residual = {:.4f})'.format(rms))

pylab.plot(x, y, 'o', color='k')
pylab.plot(x, pfit(x), color='k')
pylab.xlabel('Morti cumulativi')
pylab.ylabel('Incremento giornaliero')
pylab.show()
```
